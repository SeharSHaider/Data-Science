**Question 8c:** Identify the outliers for each team in a list format containing player names. Note: Use `to_list()` function to convert into a list format. Store your answers in `BOS_outliers` , `GSW_outliers` and `MIA_outliers`

```python
# Function to identify outliers based on IQR
def find_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers['playerName'].to_list()

# Filtering the data for each team
BOS_data = team_data[team_data['teamAbbreviation'] == 'BOS']
GSW_data = team_data[team_data['teamAbbreviation'] == 'GSW']
MIA_data = team_data[team_data['teamAbbreviation'] == 'MIA']

# Identifying outliers for each team
BOS_outliers = find_outliers(BOS_data, 'points')
GSW_outliers = find_outliers(GSW_data, 'points')
MIA_outliers = find_outliers(MIA_data, 'points')

# Displaying the outlier lists
BOS_outliers, GSW_outliers, MIA_outliers
```

**Question 12a:** Analyze the trends in player age and height over multiple seasons.

Create a dual-axis line plot that displays the average age of players and their average height for each season.

```python
nba_data['season'].unique()
```

```python
# Calculating the average age and height per season
average_stats = nba_data.groupby('season').agg({'age': 'mean', 'playerHeight': 'mean'}).reset_index()

fig, ax1 = plt.subplots(figsize=(12, 6))

# Plotting average age on the first y-axis
sns.lineplot(data=average_stats, x='season', y='age', ax=ax1, marker='o', color='blue', label='Average Age')
ax1.set_ylabel('Average Age', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Creating a second y-axis for average height
ax2 = ax1.twinx()
sns.lineplot(data=average_stats, x='season', y='playerHeight', ax=ax2, marker='o', color='orange', label='Average Height')
ax2.set_ylabel('Average Height (inches)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Adding titles and labels
plt.title('Trends in Average Player Age and Height Over Multiple Seasons')
ax1.set_xlabel('Season')

# Improving x-axis readability:
plt.xticks(rotation=45)  # Rotate the x-axis labels to prevent overlapping
ax1.set_xticks(ax1.get_xticks()[::2])  # Show every second season 


# Adding legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Display the dual-axis line plot
plt.show()
```
